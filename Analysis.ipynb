```python
from matplotlib import pyplot as plt
import numpy as np

#    
plt.title("Sample scan")
plt.xlabel("X")
plt.ylabel("Y")
print("Enter the number of pic:")
j = int(input())
image_1 = plt.imread(f'1111/Substack (38-196){j:04d}.bmp')
plt.imshow(image_1, cmap = 'gray')
plt.show()

#    
image_1 = 1 - image_1 // 255
print('Porosity of a given cut:', image_1.mean())

#-----------------------------------------------------------

size = image_1.shape

#        
print('Enter the amount of pics:')
DATA_NUM = int(input())

data = np.zeros((DATA_NUM, size[0], size[1]))

for i in range(DATA_NUM):
    image = plt.imread(f'1111/Substack (38-196){i:04d}.bmp')
    data[i] = image


#         
data = (data != 255) * 1
print('total porosity: ', data.mean())

plt.title("Sample 1")
plt.plot(range(DATA_NUM), data.mean(axis = (1, 2)), label = ' x')
plt.plot(range(size[0]), data.mean(axis = (0, 2)), label = ' y')
plt.plot(range(size[1]), data.mean(axis = (0, 1)), label = ' z')
plt.axhline(y = data.mean(), color = 'red', label = f'     {data.mean():.4f}')
plt.grid()
plt.legend()
#plt.savefig('sample_1.png')
```
